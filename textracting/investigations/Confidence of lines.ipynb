# Confidence of lines
of two noisy reports, and two non-noisy reports.

This notebook can be used to generate graphs of line confidences of any reports, provided their numbers and index in the list are modified.

```python
import pandas as pd
import json
import settings

reports = ['24362', '32676', '21838', '28298']
```

```python
i = 3
```

```python
f = json.load(open(settings.get_pagelineinfo_file(reports[i])))
```

```python
y = []
text = []
for pages in f.items():
    for page in pages[1]:
        for line in page:
            c = line['Confidence']
            t = line['Text']
            y.append(c)
            text.append(t)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (130,20)
plt.scatter(text, y)
#locs, labels = plt.xticks()
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.savefig(reports[i] + '_all_confidences.png')
plt.show()
```
